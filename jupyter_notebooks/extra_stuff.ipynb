```python
nrpts = 50
xs = np.linspace(-2, 2, num = nrpts)
ys = np.linspace(-2, 2, num = nrpts)

h = 1. / nrpts
xsf = np.meshgrid(xs, ys)[0]
ysf = np.meshgrid(xs, ys)[1]


f = (3*xsf*xsf*xsf*xsf - 4*xsf*xsf*xsf - 12*xsf*xsf + 12) / (12*(1+ysf*ysf))

#plotting as heat map
heatmap = f
heatmap = heatmap / np.max(np.max(heatmap))

print(heatmap)
#'RdBu'
cmap = plt.get_cmap('YlGnBu')
fig, ax = plt.subplots()
xsf = np.meshgrid(xs, ys)[0]
ysf = np.meshgrid(xs, ys)[1]
im = ax.pcolormesh(xsf, ysf, heatmap, cmap=cmap )
fig.colorbar(im, ax=ax)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as pyp

def line(a, b, c, x):
    return a*x*x + b*x + c

a = np.random.random()+0.5
b = 20*np.random.random()-10
c = 20*np.random.random()-10
print( a, b, c )
nrpts = 30
pts = np.zeros( (nrpts, 2))

for i in range(nrpts):
    x = np.random.random() * 10 - 5
    y = line(a,b,c, x) + 2*np.random.random()#np.random.normal(0, 30/np.max([abs(x),2]))
    pts[i,0] = x
    pts[i,1] = y

X = pts[:,0]
Y = pts[:,1]

pyp.scatter(X, Y, s=10, c='red')
pyp.show()
```
