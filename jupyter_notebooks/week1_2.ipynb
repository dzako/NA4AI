```python
#assignment 2 from the lab
M = np.array([[1, 2, 3],
             [2, 1, 2],
             [3, 2, 1]])
b = np.array([2, 3, 3])
#create copies of matrices
M0 = M.copy()
M1 = M.copy()
M2 = M.copy()
#Cramer's method
M0[:, 0] = b
M1[:, 1] = b
M2[:, 2] = b


x = [ np.linalg.det(M0)/np.linalg.det(M), np.linalg.det(M1)/np.linalg.det(M), np.linalg.det(M2)/np.linalg.det(M)]
print(x)
print(M.dot(x)) #that's exactly b
```

```python
import numpy as np
#brief introduction to pagerank algorithm
L = np.array([[0, 0.5, 0, 0], [1/3, 0, 0, 0.5], [1/3, 0, 0, 0.5], [1/3, 0.5, 1, 0]])
# initial vector of ranks r
#r = np.array([[0.25], [0.25], [0.25], [0.25]])
r = np.random.random((4,))
print("several iteration of page-rank algorithm fiding ranks of some webpages")
for i in range(50):
    r = L.dot(r)
    r = r / np.linalg.norm(r)
    print(r.T)

#r converges to the largest eigenvector 
w, v = np.linalg.eig(L)
#print the largest eigenvector (scaled by some scalar)
print("the largest eigenvector computed by numpy (that's what page-rank finds but scaled):")
print(v[:,0])
print("ratio of numpy eigv by page rank eigv (if all numbers are close the vectors are likely from the same eigenspace):")
print(v[:,0] / r.T)
```

```python
import matplotlib.pyplot as pyp
# two dimensional simple linear regression
#choose randomly bunch of points around some prescribed line
def line(a, x):
    return a*x

a = 10*np.random.random()-5
#b = 10*np.random.random()-5
print("a=%f" % a )
nrpts = 50
pts = np.zeros( (nrpts, 2))

for i in range(nrpts):
    x = np.random.random() * 20 - 10
    y = line(a,x) + np.random.normal(0, 5)
    pts[i,0] = x
    pts[i,1] = y

X = pts[:,0]
Y = pts[:,1]

pyp.scatter(X, Y, s=10, c='red')
pyp.show()
```

```python
#compute linear regression line that the best approximates the data
#the coefficient compute by solving the following analytic equation
#
#LABWORK - generalize to higher dimension
#
r = (1 / X.T.dot(X)) * X.T.dot(Y)
print(r)

pyp.scatter(X, Y, s=10, c='red')
pyp.plot(X, r*X)
pyp.show()
```

```python
#compute the derivative of sin(x) for x = pi/4 better and better approximations 
x = np.pi / 4
for prec in range(10, 200, 10):
    h = 1/prec
    der = ( np.sin(x + h)-np.sin(x) ) / h
    print("d(sin) approx using h=%f is %f" % (h, der))
```

```python
#compute the numerical derivative of sin(x) for a grid of points in the interval [0,2pi] 
#and  compare it with the graph of cos(x)
#grid of points
h = 0.001#step
xs = np.array(range(0,(int)(1/h))) * 2 * math.pi * h
ys = [ (np.sin(x+h) - np.sin(x) )/h  for x in xs ] #use list comprehension

plt.subplot(2, 1, 1)
pyp.title('numerical derivative of sin(x)')
pyp.plot(xs, ys)

plt.subplot(2, 1, 2)
pyp.title('cos(x)')
pyp.plot(xs, np.cos(xs))

pyp.show()
```

```python
#compute the numerical derivative of exp(x) for a grid of points in the interval [0,2pi] 
#and  compare it with the graph of exp(x)
#grid of points
h = 0.001#step
xs = np.array(range(0,(int)(1/h)))  * h
ys = [ (np.exp(x+h) - np.exp(x) )/h  for x in xs ] #use list comprehension

plt.subplot(2, 1, 1)
pyp.title('numerical derivative of exp(x)')
pyp.plot(xs, ys)

plt.subplot(2, 1, 2)
pyp.title('exp(x)')
pyp.plot(xs, np.exp(xs))

pyp.show()
```

```python
#compute the numerical derivative of arctan(x) for a grid of points in the interval [-1, 1] 
#and  compare it with the graph of exp(x)
#grid of points
h = 0.001#step
xs = -1 + np.array(range(0,(int)(1/h))) * 2 * h
ys = [ (np.arctan(x+h) - np.arctan(x) )/h  for x in xs ] #use list comprehension

plt.subplot(2, 1, 1)
pyp.title('numerical derivative of arctan(x)')
pyp.plot(xs, ys)

plt.subplot(2, 1, 2)
pyp.title('1/ (1+x^2)')
pyp.plot(xs, 1/ (1 + xs*xs))

pyp.show()
```

```python
#compute the numerical derivative of arctan(x) for a grid of points in the interval [-1, 1] 
#and  compare it with the graph of exp(x)
#grid of points
h = 0.001#step
xs = -1 + np.array(range(0,(int)(1/h))) * 2 * h
ys = [ (np.arctan( (x+h)*(x+h)*(x+h) ) - np.arctan(x*x*x) )/h  for x in xs ] #use list comprehension

plt.subplot(2, 1, 1)
pyp.title('numerical derivative of arctan(x^3)')
pyp.plot(xs, ys)

plt.subplot(2, 1, 2)
pyp.title('(3x^2)/ (1+x^6)')
pyp.plot(xs, (3*xs*xs)/ (1 + xs*xs*xs*xs*xs*xs))

pyp.show()
```
